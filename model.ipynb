```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import dataset
df = pd.read_csv('student_info.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
g = sns.pairplot(df)


g.map_upper(sns.kdeplot, levels=4, color=".2")

plt.show()
#pair plot for all features
```

```python
plt.scatter(x = df.study_hours , y=df.student_marks)
plt.title("Student performance")
plt.xlabel("Student Study Hours")
plt.ylabel("Student Marks")
plt.show()
#scatter plot for all features
```

```python
df.mean()
```

```python
df = df.fillna(df.mean())
df.isnull().sum()
```

```python
X = df.drop(columns = 'student_marks')
y = df.drop(columns = 'study_hours')
X.shape , y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X, y , random_state=51 , test_size=0.2)
X_train.shape , y_train.shape , X_test.shape , y_test.shape
#splitting dataset into training and testing data
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_train , y_train)
#prediction model using linear regression
```

```python
lr.score(X_test , y_test)
```

```python
lr.intercept_
```

```python
pred = lr.predict(X_test)
pred
```

```python
y_test
```

```python
pd.DataFrame(np.c_[X_test , y_test , pred] ,columns =[ 'Study hours' , 'Original Marks' , 'Predicted Marks'])
#predicted marks of all test data
```

```python
# Fine Tune Model
lr.score(X_test,y_test)
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train ,lr.predict(X_train) , color='r')
```

```python
import joblib
#saving machine learning model
```

```python
joblib.dump(lr , 'Student_Marks_Prediction_Model.pkl')
```
